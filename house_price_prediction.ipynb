```python
df.drop(columns = ['area_type' , 'availability' , 'society' , 'balcony'] , inplace = True)
```

```python
df['location'] = df['location'].fillna('Sarjapur  Road')
```

```python
df['size'] = df['size'].fillna('3 BHK')
```

```python
df['bath'] = df['bath'].fillna(df['bath'].median())
```

```python
df.info()
```

```python
df['BHK'] = df['size'].str.split(' ').str.get(0).astype('int')
```

```python
df['total_sqft'].unique()
```

```python
def convert_range(x):
  temp = x.split('-')
  if len(temp) == 2:
    return (float(temp[0]) + float(temp[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(convert_range)
```

```python
df.head()
```

```python
df['price_per_sqft'] = df['price'] * 100000 / df['total_sqft']
```

```python
df.head()
```

```python
df['location'] = df['location'].apply(lambda x: x.strip())
location_counts = df['location'].value_counts()
```

```python
location_counts_less_10 = location_counts[location_counts <= 10]
location_counts_less_10
```

```python
df['location'] = df['location'].apply(lambda x: 'other' if x in location_counts_less_10 else x)
```

```python
df['location'].value_counts()
```

```python
(df['total_sqft'] / df['BHK']).describe()
```

```python
df = df[((df['total_sqft'] / df['BHK']) >= 300)]
```

```python
df.describe()
```

```python
def remove_outlier_sqft(df):
  df_output = pd.DataFrame()
  for key,subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)

    st = np.std(subdf.price_per_sqft)
    gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft < (m+st))]
    df_output = pd.concat([df_output , gen_df] , ignore_index = True)
  return df_output

df = remove_outlier_sqft(df)
df.describe()
```

```python
def bhk_outlier_remover(df):
  exclude_indices = np.array([])
  for location,location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk , bhk_df in location_df.groupby('BHK'):
      bhk_stats[bhk] = {
          'mean': np.mean(bhk_df.price_per_sqft),
          'std': np.std(bhk_df.price_per_sqft),
          'count': bhk_df.shape[0]
      }
    for bhk,bhk_df in location_df.groupby('BHK'):
      stats = bhk_stats.get(bhk-1)
      if stats and stats['count'] > 5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
  return df.drop(exclude_indices , axis = 'index')
```

```python
df = bhk_outlier_remover(df)
```

```python
df.drop(columns = ['size' , 'price_per_sqft'] , inplace = True)
```

```python
X = df.drop(columns = ['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso , Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse = False) , ['location']),
                                       remainder = 'passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans , scaler , lr)
```

```python
pipe.fit(X_train , y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test , y_pred_lr)
```

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans , scaler , lasso)
```

```python
pipe.fit(X_train , y_train)
```

```python
y_pred_l = pipe.predict(X_test)
```

```python
r2_score(y_test , y_pred_l)
```

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans , scaler , ridge)
```

```python
pipe.fit(X_train , y_train)
```

```python
y_pred_r = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_r)
```
